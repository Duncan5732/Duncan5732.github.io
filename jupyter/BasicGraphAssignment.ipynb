# Virginia Opioids Data

```python
df_opi = pd.read_csv('OpidsVA.csv')
```

```python
df_opi
```

```python
fig, ax = plt.subplots()
 
ax.scatter(df_opi['FPOO-Rate'], df_opi['MedianHouseholdIncome'], alpha=0.5, c = 'b')  
fig.suptitle('Opioid Overdoses by Median Household Income')               # Graph title
ax.xaxis.set_label_text('Opioid Overdoses per 100,000 County Residents')  # x-axis caption
ax.yaxis.set_label_text('Median Household Income')                        # y-axis caption
fig.set_size_inches(7,5)
plt.show()

# This scatterplot illustrates that as household income decreases (less affluent) the occurrences of opioid overdoses increases.
```

```python
fig, ax = plt.subplots()
 
ax.scatter(df_opi['FPOO-Rate'], df_opi['FFHO-Rate'], alpha=0.5, c = 'b')  
fig.suptitle('Opioid Overdoses vs. Heroin Overdoses')                      # Graph title
ax.xaxis.set_label_text('Opioid Overdoses per 100,000 County Residents')   # x-axis caption
ax.yaxis.set_label_text('Heroin Overdoses per 100,000 County Residents')   # y-axis caption
fig.set_size_inches(7,5)
plt.show()

# This scatterplot effectively displays how many opioid overdoses occur, per heroin overdose, throughout VA counties.
```

# Presidents Data

```python
df_pres = pd.read_csv('presidents.csv')
```

```python
df_pres.head()
```

```python
data = df_pres['State'].values

#Finding the number of times presidents are born in specific states
freq = {}
for i in data:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
        
#Sorting the aforementioned dictionary by the frequency of states (descending order)      
sort_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
 
#This number could be adjusted and symbolizes the number of states appearing on the graph           
numPlot = 4
sort_freq = sort_freq[:numPlot]

#Setting the x and y variables for graphing purposes
x = [k for [k,j] in sort_freq]          # use this variable for the x-axis data
y = [j for [k,j] in sort_freq]          # use this variable for the y-axis data
xs = range(len(sort_freq))

# Creation of variables for the figure and axes
fig, ax = plt.subplots()

#Adjustments to the axes of the graph
ax.bar(xs, y, align='center', color='blue')
ax.xaxis.set_label_text("U.S. States", fontsize=25)
ax.yaxis.set_label_text("Count", fontsize=25)
ax.set_xticks(xs)
ax.set_xticklabels(x, rotation = 45)    #Putting the x-axis tickmarks at an angle
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.spines['right'].set_visible(False)   #Removes the exterior lines on the right and top
ax.spines['top'].set_visible(False)

#Adjustments to the figure of the graph
fig.suptitle("Most Popular Birth States of U.S. Presidents", fontsize=20)
fig.set_size_inches(9, 5)

#Displays the bar graph
plt.show()

# This frequency histogram displays the four most frequent states that U.S. presidents are born in (24 of the 44 -- 54.55%)
```

# U.S. Vehicle Data

```python
df_car = pd.read_csv('TOTALNSA.csv')
```

```python
df_car
```

```python
fig,ax = plt.subplots()
ax.plot(df_car['DATE'],df_car['TOTALNSA'])
fig.set_size_inches(18.5, 10.5)
fig.suptitle('U.S. Monthly Car Sales from 1976 to 2020', fontsize=28)
ax.set_xlabel('1976 to 2020', fontsize=25)
ax.set_ylabel('Total Sales for Month (in Thousands)', fontsize=25)
plt.show()

# Line graph showing the fluctuation in U.S. car sales over the 44 year timeframe.
```

```python
fig, ax = plt.subplots()

ax.boxplot(df_car['TOTALNSA'], labels=['All Vehicles'])
ax.yaxis.axes.set_ylim(500, 2000)
ax.set_ylabel('Total Sales for Month (in Thousands)')
fig.suptitle('Variance of U.S. Car Sales from 1976 to 2020')
fig.set_size_inches(12,6)

# This boxplot displays the variance in U.S. car sales over the given period -- great variance seen by the distance between the min, max, and the 25/75th percentile. 
```
